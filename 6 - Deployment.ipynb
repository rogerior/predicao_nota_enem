### Utilizar o melhor modelo para realizar a predição dos dados de test

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import joblib
import glob
import numpy as np
```

```python
dados = pd.read_csv('dados/test_tratado-3_2.csv', sep=';')
dados.head()
```

```python
X = dados.drop(columns=['NU_INSCRICAO', 'NU_NOTA_MT'])
X.head()
```

```python
caminho = glob.glob('modelos/melhor/Gr*')[0]
caminho
```

```python
modelo = joblib.load(caminho)
```

```python
predict = modelo.predict(X)
predict
```

```python
dados['NU_NOTA_MT'] = np.around(predict, 1)
dados[['NU_INSCRICAO', 'NU_NOTA_MT']].head(10)
```
